```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import io

# Sample data
names = ["Alice", "Bob", "Charlie", "David", "Emma"]
values = [10, 15, 7, 12, 20]
x = np.arange(len(names))
curve_y = np.interp(np.linspace(0, len(names) - 1, 100), x, values)

# Create figure and axis
fig, ax = plt.subplots()
ax.bar(names, values, color="skyblue", label="Bar")
ax.plot(
    np.linspace(0, len(names) - 1, 100),
    curve_y,
    color="red",
    linestyle="-",
    label="Curve",
)
ax.set_ylabel("Values")
ax.set_title("Bar Graph with Curve")
ax.legend()

# Convert to PIL image
buf = io.BytesIO()
plt.savefig(buf, format="png")
buf.seek(0)
pil_image = Image.open(buf)
```

```python
from ipysketch_lite import AnnotationSketch

sketch = AnnotationSketch(pil_image)
```
